```python
import numpy as np
from sklearn.datasets import fetch_california_housing
import tensorflow as tf
```

```python
housing = fetch_california_housing()
m, n = housing.data.shape
housing_plus_bias = np.c_[np.ones((m, 1)), housing.data]
m, n
```

```python
X = tf.constant(housing_plus_bias, dtype=tf.float32, name="X")
y = tf.constant(housing.target.reshape(-1, 1), dtype=tf.float32, name="y")
```

```python
n_epochs = 50
learning_rate = 0.01

theta = tf.Variable(tf.random_uniform([n + 1, 1], -1.0, 1.0), name="theta")

y_pred = tf.matmul(X, theta, name="predictions")

error = y_pred - y

mse = tf.reduce_mean(tf.square(error), name="mse")

# optimizer = tf.train.GradientDescentOptimizer(learning_rate=learning_rate)

# training_op = optimizer.minimize(mse)

gradients = (tf.matmul(tf.transpose(X), error)/m)*2

training_op = tf.assign(theta, theta - learning_rate * gradients)
```

```python
init = tf.global_variables_initializer()
with tf.Session() as sess:
    sess.run(init)
    for epoch in range(n_epochs):
        #if(epoch %100 == 0):
        #print "Epoch", epoch
        print "Error", sess.run(error)
        #print "MSE", sess.run(mse)
        #print "Theta", theta.eval()
        #print "gradients", sess.run((tf.matmul(tf.transpose(X), error)/m)*2)
        print "gradients\n",sess.run(gradients)
        #print type(gradients)
        #print gradients.dtype
        #print "learning rate", learning_rate
        #print "Multiplied by learning rate\n", sess.run(gradients * learning_rate)
        print("Epoch", epoch, "MSE =", mse.eval())
        sess.run(training_op)
    best_theta = theta.eval()
    print best_theta
    
```
